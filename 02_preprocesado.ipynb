```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import OneHotEncoder, LabelEncoder, StandardScaler
from sklearn.impute import SimpleImputer

df = pd.read_csv('train.csv')

df = df.drop(columns=["ID", "PERIODO"], errors='ignore')

print("Valores nulos por columna:\n", df.isnull().sum())

num_cols = df.select_dtypes(include=['float64', 'int64']).columns
imputer_num = SimpleImputer(strategy='mean')
df[num_cols] = imputer_num.fit_transform(df[num_cols])

cat_cols = df.select_dtypes(include=['object']).columns.drop('RENDIMIENTO_GLOBAL')
imputer_cat = SimpleImputer(strategy='most_frequent')
df[cat_cols] = imputer_cat.fit_transform(df[cat_cols])

le = LabelEncoder()
df['RENDIMIENTO_GLOBAL'] = le.fit_transform(df['RENDIMIENTO_GLOBAL'])

df = pd.get_dummies(df, columns=cat_cols, drop_first=True)

scaler = StandardScaler()
df[num_cols] = scaler.fit_transform(df[num_cols])

X = df.drop(columns=['RENDIMIENTO_GLOBAL'])
y = df['RENDIMIENTO_GLOBAL']

print("Forma final de X:", X.shape)
print("Forma final de y:", y.shape)
```

```python

preprocessed_df = pd.concat([X, y], axis=1)
preprocessed_df.to_csv('train_preprocesado.csv', index=False)

files.download('train_preprocesado.csv')

```
