```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler
from google.colab import files
uploaded = files.upload()
```

```python

df = pd.read_csv('train.csv')

print(f"Dimensiones del dataset: {df.shape}")

print("\nInformación general del dataset:")
df.info()

print("\nResumen estadístico:")
print(df.describe())

print("\nValores nulos por columna:")
print(df.isnull().sum())

df['timestamp'] = pd.to_datetime(df['timestamp'])

df['hour'] = df['timestamp'].dt.hour
df['day'] = df['timestamp'].dt.day
df['month'] = df['timestamp'].dt.month
df['weekday'] = df['timestamp'].dt.weekday

df['meter_reading_log'] = np.log1p(df['meter_reading'])

df = pd.get_dummies(df, columns=['meter'], prefix='meter')

df = df.drop(columns=['timestamp'])

df.to_csv('train_limpio.csv', index=False)
print("\nArchivo 'train_limpio.csv' guardado exitosamente.")

df_clean = pd.read_csv('train_limpio.csv')

print("\nInformación del archivo limpio:")
df_clean.info()

print("\nPrimeras filas del archivo limpio:")
print(df_clean.head())
```
